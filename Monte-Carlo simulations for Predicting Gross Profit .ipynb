He is convinced that Cogs(cost of goods sold) will be near 40% of the revenues, and their expected deviation is 20% of its own value.

### 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Define revenue mean and standard deviation from historical data

#### Use NumPy‚Äôs random.random function to simulate the potential revenue stream for 250 iterations (which is the number of trading days in a year) and then the predicted Cogs value. 

```python
rev_m = 200 # revenue mean 
rev_stdev = 10 # revenue standard deviation
iterations = 250 # 250 session for an year
```

### 3. Generate 'iterations' random values following normal distribution with rev_m as distribution mean and rev_stdev as distribution standard deviation

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

### 4. Plot the obtained revenue values in 'rev'

```python
plt.figure(figsize=(10, 6))
plt.plot(rev)
plt.xlabel('iteration')
plt.ylabel('revenue')
plt.show()
```

### To calculate gross profit, we deduct cost of goods sold(COGS) from revenues.
#### Revenues - Cost of goods sold = Gross profit
#### Gross profit - Operating expense = Operating profit

```python
COGS = - (rev * np.random.normal(0.4,0.2)) # cogs is 40% of revenue with 20% standard deviation. COGS is negative as this is cost incurred by company
COGS
```

### 5.  Plot the obtained COGS values in 'COGS'

```python
plt.figure(figsize=(10, 6))
plt.plot(COGS)
plt.xlabel('iteration')
plt.ylabel('cogs')
plt.show()
```

### 6. Evaluate mean of all the COGS values obtained 

```python
COGS.mean()
```

### 7. Evaluate standard deviation of all the COGS values obtained

```python
COGS.std()
```

### 8. Evaluate the Gross Profit 

```python
Gross_Profit = rev + COGS # we have already taken cogs as negative
Gross_Profit
```

### 9. Plot the obtained Gross profit values

```python
plt.figure(figsize=(10, 6))
plt.plot(Gross_Profit)
plt.xlabel('iteration')
plt.ylabel('gross profit')
plt.show()
```
